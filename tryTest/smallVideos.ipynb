```python
import re
import requests
```

```python
def getUrls():
    url = "https://www.kk11vv.com:2083"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.157 Safari/537.36'
    }
    response = requests.get(url, headers = headers)
    response.encoding = 'utf-8'
    res = str(response.text).replace(' ', '')
    string = "<ahref=\"(/html/[0-9]+?-[0-9]+?.html)\"target=\"_blank\"><img"
    urls = re.compile(string).findall(res)
    return urls
```

```python
def getVideo(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.157 Safari/537.36'
    }
    response = requests.get(url, headers = headers)
    response.encoding = 'utf-8'
    #print(response.text.replace(' ', ''))
    res = str(response.text).replace(' ', '')
    string = 'IFRAMEmarginWidth=0marginHeight=0src="(/public/player/.*?/m3u8.m3u8)"frameBorder=0'
    currUrl = re.compile(string).findall(res)
    myFile = open("./videos/list.txt", "a", encoding='utf-8')
    myFile.write(str("https://www.kk11vv.com:2083"+currUrl[0]+'\n'))
    myFile.close()
    print("https://www.kk11vv.com:2083"+currUrl[0])
    
```

```python

def getVideos():
    url = "https://www.kk11vv.com:2083"
    urls = getUrls()
    for currUrl in urls:
        getVideo(url + currUrl)

getVideos()
```

```python
import os
def loadTs():
    root = "./video/"
    try:
        if not os.path.exists(root):
            os.mkdir(root)
        for i in range(889):
            path = root + str(i).zfill(4) + ".ts"
            r = requests.get("https://vod.se0273.com//2019/91cf0fa2/" + str(i).zfill(4) + ".ts")
            r.encoding = 'utf-8'
            with open(path, 'w') as f:
                f.write(str(r.content))
                f.close()
                print(str(i).zfill(4) + ".ts 完成")
        print("全部ts下载完毕")
    except Exception as e:
        print("批量下载失败"+str(e))
```
