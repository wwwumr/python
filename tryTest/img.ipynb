```python
import re
import requests
```

```python
url = 'https://www.58pic.com/piccate/53-0-0.html?tid=16540'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.157 Safari/537.36"}
res = requests.get(url, headers = headers)
res.encoding = 'utf-8'
resText = str(res.text).replace(' ', '')
print(resText)
```

```python
reString = '<imgsrc="(//preview.*?)"alt'
results = re.compile(reString).findall(resText)
print(results)
#<img src="//preview.qiantucdn.com/58pic/35/03/56/27t58PICFe0wh6cR6Vwe3_PIC2018.jpg!qt324new_nowater" alt="蓝天天空韩国街道风景城市景色旅游旅行摄影" data-i="0" class="back-static" width="324" height-no="243">
```

```python
#<img class="back-static" data-original="//preview.qiantucdn.com/58pic/35/05/85/43E58PICXbMczk4rF7sUH_PIC2018.jpg!qt324new_nowater" src="//preview.qiantucdn.com/58pic/35/05/85/43E58PICXbMczk4rF7sUH_PIC2018.jpg!qt324new_nowater" title="现代建筑科技商业大厦" alt="现代建筑科技商业大厦" width="324" height-no="216" style="display: block;">
nextString = '<imgclass=".*?"data-original="(//preview.*?)"'
results += re.compile(nextString).findall(resText)
print(results)
```

```python
for i in range(len(results)):
    try:
        response = requests.get("https:" + results[i], headers = headers)
        fwriter = open("./img/"+ str(i+1).zfill(2) + ".jpg", "wb")
        fwriter.write(response.content)
        fwriter.close()
        print(results[i] + " 完成")
    except Exception as e:
        print(str(e))
```
